```python
import MMM_010519 as MMM
import warnings
import pandas as pd
import numpy as np
import os
warnings.filterwarnings("ignore", message="divide by zero encountered in log")
```

```python
cross_val_datapath = '/Users/dror/PycharmProjects/MMM workshop/data/samples_cv_data/'
num_of_folds = 10

training_sets = sorted([file for file in os.listdir(cross_val_datapath) if file.startswith('90%')])
test_sets = sorted([file for file in os.listdir(cross_val_datapath) if file.startswith('10%')])

# for i in range(n):
    
```

First, Fit two new denovo MMM models to each training set

```python

MMM.set_num_of_topics(12)
MMM.set_num_of_samples(int(560*0.9))
models_dict = {}


for i in range(num_of_folds):
    MMM.set_num_of_topics(12)
    mut_counts = pd.read_csv(cross_val_datapath+training_sets[i], index_col = 0)
    model = model = MMM.fit(mut_counts, 0.001, 50)
    e = np.exp(pd.DataFrame(model[0].e)).values
    name = f'MMM_{MMM.num_of_topics}_e_{i+1}'
    models_dict[name] = e
    
    MMM.set_num_of_topics(14)
    model = model = MMM.fit(mut_counts, 0.001, 50)
    e = np.exp(pd.DataFrame(model[0].e)).values
    name = f'MMM_{MMM.num_of_topics}_e_{i+1}'
    models_dict[name] = e    
    
```

Gather all matrices from the relevant folder, they are numbered according to the different data set id

```python

lda12_mats = [file for file in os.listdir(cross_val_datapath) if file.startswith('LDA_12')]
lda14_mats = [file for file in os.listdir(cross_val_datapath) if file.startswith('LDA_14')]
ctm12_mats = [file for file in os.listdir(cross_val_datapath) if file.startswith('CTM_12')]
ctm14_mats = [file for file in os.listdir(cross_val_datapath) if file.startswith('CTM_14')]

all_mats_paths = lda12_mats + lda14_mats + ctm12_mats + ctm14_mats
```

```python
# for mat in lda12_mats:
#     e = pd.read_csv(cross_val_datapath+mat,  index_col=0).values
     
```

Create model matrices dictionary based on their name

```python
for mat in all_mats_paths:
    e = pd.read_csv(cross_val_datapath+mat,  index_col=0).values
    if(mat.startswith('CTM')):
        e = pd.read_csv(cross_val_datapath+mat,  index_col=0).reset_index().values
        e = np.exp(e)
        id = mat.split('_')[3].split('.')[0]
    
        
    else:
        id = mat.split('_')[5].split('.')[0]
        
    name = f'{mat[0:6]}_e_{id}'
    models_dict[name] = e
```

```python
len(models_dict.keys()) #6 models, 10 folds -> Should be 60
```

```python
models_by_set = {}
for i in range(1, num_of_folds+1):
    models_by_set[f'set_{i}'] = [model for model in models_dict.keys() if model.endswith(f'{i}')]
    
    
```

Refit each model  and apply it to the test set with the correct id

```python
fixed_e_flag = True
MMM.set_num_of_samples(int(560*0.1))

df = pd.DataFrame(columns = ['MMM_12', 'MMM_14', 'LDA_12', 'LDA_14', 'CTM_12', 'CTM_14'])

for i in range(1, num_of_folds+1):
# for i in range(1, 3):    
    data = pd.read_csv(cross_val_datapath+test_sets[i-1], index_col = 0)
    models = models_by_set[f'set_{i}']
    ll_vec = []
    print(i)
    for model in models:
        MMM.set_num_of_topics(int(model.split('_')[1]))
        mat = models_dict[model]
        refitted_model = MMM.fit(data, 0.001, 50, mat, None, fixed_e_flag, False )
        ll = MMM.Theta.log_likelihood2(data, refitted_model)
        ll_vec.append(ll)
    
#     ll_vec = pd.Series(ll_vec)
#     print(ll_vec)
#     df.loc[i] = pd.Series(ll_vec)
    df.loc[i] = ll_vec

```

Add a summary of all folds, transfer to long form

```python
df_final = df.T
df_final = df_final.assign(score = df_final.sum(axis = 1))
```

```python
df_final.to_csv("data/final_data/all_models_samples_crossval.csv")
```
